# **Introdução**

Este projeto se insere no contexto da mudança global do clima, com ênfase nas interações entre o clima e os ecossistemas terrestres, em especial a biodiversidade. A área de conhecimento envolvida é multidisciplinar, abrangendo ciências ambientais, climatologia e sustentabilidade, as quais são essenciais para compreender as interações entre as atividades humanas e o meio ambiente. O foco é na "ODS 15 - Vida Terrestre", que visa proteger, restaurar e promover o uso sustentável dos ecossistemas terrestres, gerenciando de forma sustentável as florestas e combatendo a desertificação e a perda de biodiversidade (ONU, 2015).
A mudança global do clima, caracterizada por alterações nos padrões climáticos globais, é um dos fenômenos mais desafiadores e amplamente discutidos no século XXI. Este processo resulta de interações complexas entre fatores naturais e antrópicos, com destaque para a emissão de gases de efeito estufa (GEE) na atmosfera, principalmente derivados da queima de combustíveis fósseis e mudanças no uso da terra, como o desmatamento. Essas alterações impactam diretamente o equilíbrio ecológico e o bem-estar humano, provocando eventos climáticos extremos, como secas severas, enchentes, elevação do nível do mar e mudanças nos ecossistemas. O entendimento das dinâmicas climáticas e suas consequências torna-se fundamental para a formulação de políticas públicas eficazes e a implementação de práticas sustentáveis que mitiguem esses efeitos e promovam a adaptação das populações vulneráveis (IPCC, 2021).
O problema selecionado para análise refere-se às alterações na biodiversidade decorrentes das mudanças climáticas no Brasil, em especial nas florestas tropicais, como a Amazônia e a Mata Atlântica. Essas áreas são essenciais para a regulação do clima global e abrigam uma vasta gama de espécies que são diretamente afetadas pelas mudanças no clima. As mudanças no padrão de precipitação e aumento das temperaturas já resultam em perdas significativas de biodiversidade, com implicações para os serviços ecossistêmicos essenciais para a sobrevivência humana, como a purificação da água, a fertilidade do solo e a captura de carbono (PEREIRA et al., 2010).
Através deste projeto, busca-se analisar os efeitos da mudança climática sobre a biodiversidade terrestre no Brasil e apresentar soluções e estratégias que possam ser adotadas tanto em nível local quanto global para mitigar esses efeitos.

# **Descrição da base de dados**

IMF - Annual Surface Temperature Change:
Esse conjunto de dados apresenta estimativas anuais das mudanças na temperatura média da superfície, em relação à climatologia de referência do período de 1951-1980. As temperaturas são medidas em graus Celsius e estão disponíveis de 1961 a 2021, cobrindo países individualmente e o mundo como um todo.
A fonte dos dados é a Organização das Nações Unidas para Alimentação e Agricultura (FAO), e o conjunto está estruturado em períodos anuais por país, permitindo a análise das tendências de mudanças climáticas globais e regionais ao longo do tempo​.
https://climatedata.imf.org/datasets/4063314923d74187be9596f10d034914_0/explore

Our World in Data - Climate Change:
Essa plataforma oferece uma ampla variedade de indicadores relacionados às mudanças climáticas. A base de dados inclui medidas como emissões de gases de efeito estufa, temperatura global, níveis de poluição do ar, e outros indicadores de impacto ambiental.
A estrutura dos dados é organizada por países e regiões, permitindo uma análise detalhada do progresso em diferentes áreas ao longo de várias décadas.
A plataforma disponibiliza séries temporais de dados, gráficos e análises, com um período de coleta abrangendo várias décadas, dependendo do indicador.
https://ourworldindata.org/climate-change

INMET - Dados Históricos:
O Instituto Nacional de Meteorologia (INMET) disponibiliza dados históricos sobre condições climáticas no Brasil. As informações incluem variáveis como temperatura, umidade, precipitação e velocidade do vento, coletadas de estações meteorológicas em todo o país.
A estrutura dos dados é organizada em planilhas por estação meteorológica e por períodos de coleta que podem abranger vários anos. Esses dados são atualizados regularmente e podem ser baixados em diferentes formatos, permitindo análises detalhadas por região e período.
https://portal.inmet.gov.br/dadoshistoricos

MapBiomas - Coleções MapBiomas:
O MapBiomas fornece séries temporais sobre o uso e cobertura da terra no Brasil, com foco em mudanças nos ecossistemas e biomas ao longo dos anos.
As coleções são divididas por temas como vegetação nativa, agropecuária, e áreas urbanas, e estão disponíveis em mapas de alta resolução, organizados por anos e biomas.
O período de coleta começa em 1985 e se estende até os dias atuais, permitindo análises detalhadas da evolução ambiental no Brasil.
https://brasil.mapbiomas.org/colecoes-mapbiomas/

CEMADEN - Monitoramento de Seca no Brasil:
Esse conjunto de dados do CEMADEN fornece informações detalhadas sobre o monitoramento de secas no Brasil, incluindo análises de impactos socioeconômicos e ambientais das secas.
Os dados são coletados regularmente desde 2013 e incluem mapas e relatórios mensais, organizados por estados e municípios brasileiros.
A estrutura é geoespacial, com indicadores que permitem visualizar as áreas mais afetadas e as variações temporais das secas no país.
https://www.gov.br/cemaden/pt-br/assuntos/monitoramento/monitoramento-de-seca-para-o-brasil/monitoramento-de-secas-e-impactos-no-brasil-2013-setembro-2023-1
